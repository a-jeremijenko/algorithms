```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = ''.join(filter(str.isalnum, s)).lower()
        return True if s == s[::-1] else False
    
Solution().isPalindrome("A man, a plan, a canal: Panama")
```

## Similar string chain

```python
class Solution:
    def numSimilarGroups(self, A: List[str]) -> int:
        A = list(set(A))                          #
        n, m = len(A), len(A[0])
        self.p = list(range(n))                    #
        self.nnm(A)
        return len(set(map(self.uni, self.p)))    #
    
    def uni(self, x: int):                      #
        if x != self.p[x]:
            self.p[x] = self.uni(self.p[x])
        return self.p[x]
        
    def nnm(self, A: List[str]):                #O(N^2*M)
        n, m = len(A), len(A[0])
        def check(x, y):                        #
            t = 0
            for i in range(m):
                if x[i] != y[i]:
                    t += 1
                    if t > 2:
                        return False
            return True
        for i in range(n):
            for j in range(i + 1, n):           #
                pi, pj = self.uni(i), self.uni(j)
                if pi != pj and check(A[i], A[j]):
                    self.p[pj] = pi

strs = ["tars","rats","arts","star","star"]
Solution().numSimilarGroups(strs)
```
