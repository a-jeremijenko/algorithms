```python
import heapq
arr = [0,1,1,2,4,4,1,3,3,2]
k = 6
```

```python
class Solution:
    def getLeastNumbers(self, arr, k: int):
        return heapq.nsmallest(k, arr)
Solution().getLeastNumbers(arr, k)
```

```python
class Solution:
    def getLeastNumbers(self, arr: List[int], k: int) -> List[int]:
        if not k: return []
        hp = arr[:k]
        heapq._heapify_max(hp)
        for num in arr[k:]:
            if num < hp[0]:
                heapq._heapreplace_max(hp, num)
        return list(hp)
Solution().getLeastNumbers(arr, k)
```

```python
# Manual implementation of quick search
class Solution:
    def getLeastNumbers(self, arr: List[int], k: int) -> List[int]:
        def partition(nums, left, right): # assume  left < right
            pivot, i = nums[right], left
            for j in range(left, right): 
                if nums[j] < pivot:
                    nums[i], nums[j] = nums[j], nums[i]
                    i += 1
            nums[i], nums[right] = nums[right], nums[i]
            return i
        
        if k == 0: # 
            return []
        
        left, right, target = 0, len(arr)-1, k-1

        while left < right:
            ind = partition(arr, left, right) 
            if ind == target: 
                return arr[:target+1]  # 
            elif ind < target:  
                left = ind + 1
            else: 
                right = ind -1
        return arr[:left+1] 
Solution().getLeastNumbers(arr, k)
```

```python
arr = [
  [1,   4,  7, 11, 15],
  [2,   5,  8, 12, 19],
  [3,   6,  9, 16, 22],
  [10, 13, 14, 17, 24],
  [18, 21, 23, 26, 30]
]

class Solution:
    def findNumberIn2DArray(self, matrix: List[List[int]], target: int) -> bool:
        i, j = len(matrix) - 1, 0
        while i >= 0 and j < len(matrix[0]):
            if matrix[i][j] > target: i-=1
            elif matrix[i][j] < target: j+=1
            else: return True
        return False
Solution().findNumberIn2DArray(arr, 6)
        
```
