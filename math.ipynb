# Find the (rounded down) SQRT of X

```python
class Solution:
    def mySqrt(self, x: int) -> int:
        if x == 0:
            return 0
        else:
            t = x
            while t > x/t:
                t = (t+x/t) // 2
            return int(t)
Solution().mySqrt(16)
```

## Combinations

```python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        res = [[]]
        for i in nums:
            res = res + [[i] + num for num in res]
    return res
```

```python
Solution().subsets([1, 3, 4, 5])
```
