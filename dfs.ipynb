```python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        if not digits:
            return []
        dicts = {'2':['a','b','c'],
                 '3':['d','e','f'],
                 '4':['g','h','i'],
                 '5':['j','k','l'],
                 '6':['m','n','o'],
                 '7':['p','q','r','s'],
                 '8':['t','u','v'],
                 '9':['w','x','y','z']}
        res = []
        def getStr(strs, index, lens, res):
            if index == lens:
                res.append(strs)
                return
            strList = dicts[digits[index]]
            for i in range(len(strList)):
                temp = strs + strList[i]
                getStr(temp, index + 1, lens, res)
        getStr("", 0, len(digits), res)
        return res
myObj = Solution()
digits = "23"
myObj.letterCombinations(digits)
```
