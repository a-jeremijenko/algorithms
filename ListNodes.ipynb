```python
class Node(object):
    def __init__(self, val=0, left=None, right=None):
        self.val= val
        self.left = left
        self.right = right
    def __str__(self):
        return f'<{self.val}, {self.left}, {self.right}>'  # Py 3.6

from collections import deque
def listToBinTree(data):    
    n = iter(data)
    tree = Node(next(n))
    fringe = deque([tree])
    while True:
        head = fringe.popleft()
        try:
            if (left:=next(n)) is not None:
                head.left = Node(left)
                fringe.append(head.left)
            if (right:=next(n)) is not None:
                head.right = Node(right)
                fringe.append(head.right)
        except StopIteration:
            break
    return tree
        
class Solution:
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
        def dfs(p, q):
            if p==None and q==None: return True
            if p!=None and q!=None and p.val==q.val:
                return dfs(p.left, q.left) and dfs(p.right, q.right)
            return False
        return dfs(p, q)
    
L1 = listToBinTree([1, 2, 3])
L2 = listToBinTree([1, 2, 3])
L3 = listToBinTree([1, 2, None])
L4 = listToBinTree([1, None, 3])

print(Solution().isSameTree(L1, L2))
print(Solution().isSameTree(L3, L4))
```
